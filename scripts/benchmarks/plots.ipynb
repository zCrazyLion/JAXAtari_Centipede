## 1. Throughput comparison

```python
from pathlib import Path
import re

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid")

# Fill this list with the throughput CSV files you want to compare.
csv_paths = [
    "./outputs/throughput_results_jaxatari_final_p1_20260504_215939.csv",
    "./outputs/throughput_results_jaxatari_final_p2_20260504_220022.csv",
    "./outputs/throughput_results_jaxatari_final_p3_20260504_220400.csv",
    "./outputs/throughput_results_jaxatari_final_p4_20260504_220537.csv",
]

include_unknown_envs = False


def build_impl_label(row: pd.Series) -> str:
    backend = str(row.get("backend", "unknown")).strip().lower()

    if backend == "jaxatari":
        mode = str(row.get("jaxatari_mode", "")).strip() or "unknown"
        platform = str(row.get("jax_platform", "")).strip().lower()
        platform_part = f", {platform.upper()}" if platform else ""
        return f"JAXAtari ({mode}{platform_part})"

    if backend == "gxm":
        env_type = str(row.get("gxm_env_type", "")).strip()
        impl = str(row.get("gxm_impl", "")).strip()
        detail = ", ".join(part for part in [env_type, impl] if part)
        return f"GXM ({detail})" if detail else "GXM"

    return backend.upper() if backend else "UNKNOWN"


def extract_game_name(env_name: str) -> str:
    name = str(env_name or "").strip()
    if not name:
        return "unknown"

    name = name.split("/")[-1]
    name = re.sub(r"-v\d+$", "", name, flags=re.IGNORECASE)
    key = re.sub(r"[^a-z0-9]", "", name.lower())

    # Common alias normalization for short/legacy game tokens in logs.
    alias_map = {
        "fro": "frostbite",
        "montezuma": "montezumarevenge",
    }
    return alias_map.get(key, key) or "unknown"


def aggregate_plot_data(input_df: pd.DataFrame) -> pd.DataFrame:
    return (
        input_df.groupby(["game_name", "num_envs", "implementation"], as_index=False)["throughput_env_steps_per_sec"]
        .median()
        .sort_values(["game_name", "num_envs", "implementation"])
)


def plot_throughput_grid(data: pd.DataFrame, title: str) -> None:
    if data.empty:
        print(f"No data available for: {title}")
        return

    games = sorted(data["game_name"].unique())
    g = sns.relplot(
        data=data,
        x="num_envs",
        y="throughput_env_steps_per_sec",
        hue="implementation",
        col="game_name",
        col_order=games,
        col_wrap=3,
        kind="line",
        marker="o",
        facet_kws={"sharey": False, "sharex": True},
        height=3.4,
        aspect=1.15,
    )

    g.set_axis_labels("Number of parallel envs", "Throughput (env steps / sec)")
    g.set_titles(col_template="{col_name}")

    x_ticks = sorted(data["num_envs"].unique())
    for axis in g.axes.flatten():
        axis.set_xscale("log", base=2)
        axis.set_xticks(x_ticks)
        axis.get_xaxis().set_major_formatter(plt.ScalarFormatter())

    g.figure.subplots_adjust(right=0.82, top=0.9)
    if g._legend is not None:
        sns.move_legend(
            g,
            "center left",
            bbox_to_anchor=(1.02, 0.5),
            title="Implementation",
            frameon=True,
        )

    g.figure.suptitle(title, y=0.98)
    plt.show()


resolved_paths = [Path(path).expanduser() for path in csv_paths]

missing_paths = [path for path in resolved_paths if not path.exists()]
if missing_paths:
    raise FileNotFoundError("Missing CSV file(s):\n" + "\n".join(str(path) for path in missing_paths))

frames = []
for path in resolved_paths:
    frame = pd.read_csv(path)
    frame["source_file"] = str(path)
    frames.append(frame)

df = pd.concat(frames, ignore_index=True)

df = df[df["status"].fillna("ok").eq("ok")].copy()
df = df[pd.to_numeric(df["throughput_env_steps_per_sec"], errors="coerce").notna()].copy()
df["throughput_env_steps_per_sec"] = df["throughput_env_steps_per_sec"].astype(float)
df["num_envs"] = pd.to_numeric(df["num_envs"], errors="coerce")
df = df[df["num_envs"].notna()].copy()
df["num_envs"] = df["num_envs"].astype(int)

df["implementation"] = df.apply(build_impl_label, axis=1)
df["env_name"] = df["env_name"].fillna("unknown")
df["game_name"] = df["env_name"].apply(extract_game_name)

if not include_unknown_envs:
    df = df[df["game_name"] != "unknown"].copy()

plot_df = aggregate_plot_data(df)
plot_throughput_grid(plot_df, "Throughput comparison (all runs)")

is_jaxatari_oc = (
    df["backend"].fillna("").astype(str).str.lower().eq("jaxatari")
    & df["jaxatari_mode"].fillna("").astype(str).str.lower().str.contains("oc")
)
df_no_oc = df[~is_jaxatari_oc].copy()
plot_df_no_oc = aggregate_plot_data(df_no_oc)
plot_throughput_grid(plot_df_no_oc, "Throughput comparison (excluding JAXAtari OC runs)")

print("Loaded CSV files:")
for path in resolved_paths:
    print("-", path)

display(plot_df.head(20))
display(plot_df_no_oc.head(20))
```
